# E-Commerce Sales Analytics Project\n## Phase 3 — Python EDA\n

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel("//content//E-Commerce Orders.csv.xlsx")
```

```python
#Data Cleaning
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python

# Monthly Sales Trend
plt.figure(figsize=(10,5))
df.groupby(df['Date'].dt.to_period('M'))['TotalPrice'].sum().plot(kind='line')
plt.title('Monthly Sales Trend')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sales by Category
plt.figure(figsize=(10,5))
df.groupby('Product')['TotalPrice'].sum().plot(kind='bar')
plt.title('Sales by Product')
plt.show()
```

```python
# Top Products
plt.figure(figsize=(10,5))
df.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Top Products')
plt.show()
```

```python
# Revenue by Region
# Revenue by Shipping Address

plt.figure(figsize=(10,5))

df.groupby('ShippingAddress')['TotalPrice'].sum().plot(kind='bar')

plt.title('Revenue by Shipping Address')
plt.xlabel('Shipping Address')
plt.ylabel('Revenue')

plt.show()
```

```python
# Marketing Channel Performance
# Marketing Channel Performance

plt.figure(figsize=(10,5))

df.groupby('ReferralSource')['TotalPrice'].sum().plot(kind='bar')

plt.title('Marketing Channel Performance')
plt.xlabel('Referral Source')
plt.ylabel('Revenue')

plt.show()
```

```python
# Payment Method Usage
plt.figure(figsize=(10,5))
df['PaymentMethod'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Payment Method Usage')
plt.show()
```

```python
# Customer Order Frequency
plt.figure(figsize=(10,5))
df.groupby('CustomerID')['OrderID'].count().hist()
plt.title('Customer Order Frequency')
plt.show()
```

```python
# Average Order Value
plt.figure(figsize=(10,5))
df.groupby('OrderID')['TotalPrice'].sum().hist()
plt.title('Average Order Value')
plt.show()
```

```python
# Top Customers
# Top Customers

plt.figure(figsize=(10,5))

df.groupby('CustomerID')['TotalPrice']\
  .sum()\
  .sort_values(ascending=False)\
  .head(10)\
  .plot(kind='bar')

plt.title('Top Customers')
plt.xlabel('CustomerID')
plt.ylabel('Revenue')

plt.show()
```

```python
# Daily Revenue Trend
plt.figure(figsize=(10,5))
df.groupby('Date')['TotalPrice'].sum().plot(kind='line')
plt.title('Daily Revenue Trend')
plt.show()
```
